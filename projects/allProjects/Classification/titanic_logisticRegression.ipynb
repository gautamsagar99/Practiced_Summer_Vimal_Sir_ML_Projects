```python
# now X has multiple categorical variables so we do lable encoding and one-hot encoding.
```

```python
# For both operations we can use get_dummies()
```

```python
# For Pclass field
```

```python
Pclass = pd.get_dummies(final_dataset['Pclass'], dtype=int, drop_first=True)
```

```python
# Pclass
```

```python
# For Sex field
```

```python
Sex = pd.get_dummies(final_dataset['Sex'], dtype=int, drop_first=True)
```

```python
# Sex
```

```python
# For SibSp field
```

```python
SibSp = pd.get_dummies(final_dataset['SibSp'], dtype=int, drop_first=True)
```

```python
# For Parch field
```

```python
Parch = pd.get_dummies(final_dataset['Parch'], dtype=int, drop_first=True)
```

```python
# Parch
```

```python
# For Embarked field
```

```python
Embarked = pd.get_dummies(final_dataset['Embarked'], dtype=int, drop_first=True)
```

```python
complete_ds = pd.concat([Pclass, Sex, SibSp, Parch, Embarked, final_dataset['Age'], final_dataset['Survived']], axis=1)
```

```python
# X.columns = X.columns.astype(str)
```

```python
complete_ds.columns
```

```python
complete_ds.dropna()
```

```python
complete_ds
```

```python
y = complete_ds['Survived']
```

```python
y
```

```python
complete_ds = complete_ds.drop(columns='Survived')
```

```python
complete_ds
```

```python
# Now entire Preprocessing is done now we can split into trainig and test and predict the model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = complete_ds
```

```python
X
```

```python
X.columns
```

```python
X.columns =  X.columns.astype(str)
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```
